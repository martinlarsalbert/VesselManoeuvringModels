# MDL Twin simulations: Select runs
This notebooks makes a selection of runs from the MDL DB, to be used in the twin simulations

```python
#import data
from mdldb.mdl_db import MDLDataBase
from mdldb.tables import Base, Model, LoadingCondition, Project, Ship
from mdldb.run import Run
from mdldb.authenticate import authenticate
import pandas as pd
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
pd.set_option("display.max_columns", None)
import warnings
import numpy as np
```

## Get MDL runs

```python
authenticate(dotenv_path='../.env')
db = MDLDataBase()
```

```python
project = db.session.query(Project).get(40199079)
```

```python
query = db.session.query(Run,Model,LoadingCondition,Ship).filter((Run.project==project)).join(
    Model, Run.model_number == Model.model_number).join(
    LoadingCondition, Run.loading_condition_id==LoadingCondition.id).join(
    Ship, Run.ship_name==Ship.name)

df_all = pd.read_sql(query.statement, db.engine)
df_all.drop(columns=['model_number_1','model_number_2','id_1','name_1','ship_name_1'], inplace=True)
df_all.set_index('id', inplace=True)
display(df_all.head())
display(df_all.tail())
```

```python
mask = df_all['comment'].str.contains('Seglande')
df_all.loc[mask,'test_type'] = 'sailing'
```

## Not zero speed

```python
df_select = df_all.dropna(subset=['ship_speed'])
mask = df_select['ship_speed'] > 0
df_select = df_select.loc[mask].copy()
```

```python
df_select['ship_speed'].unique()
```

## test type

```python
wanted_test_types = ['reference speed',
                    'rodergrundvinkel',
                    'sailing',
                    'zigzag',
                    'turning circle']
mask = df_select['test_type'].isin(wanted_test_types)
df_select = df_select.loc[mask].copy()
```

```python
df_select['comment'].unique()
```

```python
unwanted_comments = [
    'Kal flÃ¤ktstyrning',
#    'ZZ 10/10SB - GWA= -90',
#    'ZZ 10/10PS - GWA= -90',
#    'ZZ 20/20SB - GWA= -90',
#    'ZZ 20/20PS - GWA= -90'
]
mask = df_select['comment'].isin(unwanted_comments)
df_select = df_select.loc[~mask].copy()
```

```python
df_select.describe()
```

## To model scale

```python
df_select.iloc[0].dropna()
```

```python
df_select_model_scale = df_select.copy()
scale_factor = df_select_model_scale['scale_factor']
units = {}

df_select_model_scale['ship_speed']*=1.852/3.6*1/np.sqrt(scale_factor)
units['ship_speed'] = r'm/s'

df_select_model_scale['Volume']*=1/(scale_factor**3)
units['Volume'] = r'm^3'

meters = ['lcg','kg','gm','TF','TA','KXX','KZZ','lpp','beam']
for key in meters:
    df_select_model_scale[key]*=1/scale_factor
    units[key] = r'm'
```

```python
df_select_model_scale['xm'] = 0
df_select_model_scale['ym'] = 0
df_select_model_scale['zm'] = -0.214
units['xm'] = 'm'
units['ym'] = 'm'
units['zm'] = 'm'
```

## Save

```python
df_select_model_scale.to_csv('../data/interim/run_selection.csv')
pd.Series(units).to_csv('../data/raw/units.csv')
```
